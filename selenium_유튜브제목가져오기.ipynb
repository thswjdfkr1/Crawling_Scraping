```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

from webdriver_manager.chrome import ChromeDriverManager

import bs4
import os
import time
```

```python
# 드라이버 경로 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 화면 출력 x
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage') # /deb/shm 디렉토리를 사용하지 않음 공유메모리를 담당

driver_path = r"C:\Users\jeong\Downloads\chrome-win64chromedriver.exe"
driver = webdriver.Chrome(options = options)
```

```python
# 사이트접속
driver.get('https://www.youtube.com/results?search_query=%EB%91%90%EC%82%B0%EB%B2%A0%EC%96%B4%EC%8A%A4')

time.sleep(5)

i = 0
body = driver.find_element(By.TAG_NAME, 'body')
while 1 < 10:
  body.send_keys(Keys.PAGE_DOWN)
  time.sleep(2)
  i = i + 1

# html 읽어오기
html = driver.page_source
print(html)
```
