```python
import os

# Scrapy 프로젝트 구조 만들기
os.makedirs("/content/example_project/spiders", exist_ok=True)

# quotes_spider.py 파일 생성
spider_code = """
import scrapy

class QuotesSpider(scrapy.Spider):
    name = 'example'
    start_urls = [input("크롤링할 URL을 입력하세요: ")]

    def parse(self, response):
        # 기사 제목 추출
        title = response.css('h1.title::text').get()

        # 작성자 추출
        author = response.css('span.author::text').get()

        # 기사 내용 추출
        content = response.css('div.article-body p::text').getall()

        yield {
            'title': title,
            'author': author,
            'content': ' '.join(content),
        }
"""

# 파일로 저장
with open("/content/example_project/example_project/spiders/example_spider.py", "w") as file:
    file.write(spider_code)
```

# CrawlerProcess
 - Scrapy 스파이더를 실행하고 크롤링이 끝난 후에 자동으로 종료

```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings
from scrapy import signals
from twisted.internet import defer
```

```python
ITEM_PIPELINES = {'myproject.pipelines.JsonWriterPipeline': 1}

# 크롤러 설정
process = CrawlerProcess(settings={
    'FEED_FORMAT': 'json',
    'FEED_URI': '/content/example2.json',
})

# 큐가 다 비어지면 reactor.stop()으로 종료
@defer.inlineCallbacks
def run_spider():
    yield process.crawl(QuotesSpider)
    process.start(stop_after_crawl=True)

# 스파이더 실행
run_spider()
```
