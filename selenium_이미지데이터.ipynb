```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import os

from urllib.request import urlretrieve
```

```python
if not os.path.isdir("C:\\Users\\thswj\\Desktop\\vscode\\img"):
    os.makedirs("C:\\Users\\thswj\\Desktop\\vscode\\img")
else:
    print('폴더가 이미 있어요')
                     
```

```python
driver = webdriver.Chrome()
driver.get('https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=%EB%91%90%EC%82%B0%EB%B2%A0%EC%96%B4%EC%8A%A4')

# 스크롤
for i in range(3):
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    
# 이미지 수집
img = driver.find_elements(By.CSS_SELECTOR, 'img._fe_image_tab_content_thumbnail_image')

img_list = []
for i in img:
    img_list.append(i.get_attribute('src'))
    
# 수집한 이미지 경로를 이용한 폴더내 파일로 저장
for i in range(len(img_list)):
    urlretrieve(img_list[i], f"C:\\Users\\thswj\\Desktop\\vscode\\img\\{i}.jpg")
    
driver.close()
```
