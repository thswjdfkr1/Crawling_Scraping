```python
from urllib.parse import quote
```

## URL 만들기

```python
str = input('검색어를 입력하세요. ')
keyword = quote(str)

target_url = 'https://search.hani.co.kr/search/newslist?searchword={}&sort=desc&startdate=1988.01.01&enddate=2025.04.21&dt=all'.format(keyword)
print('{}에 대한 검색결과 > '.format(str), target_url)
```

## 첫 화면의 HTML 가져오기

```python
import requests
html = requests.get(target_url).text
print(html)
```

## 검색된 기사 개수 가져오기

```python
from bs4 import BeautifulSoup

bs = BeautifulSoup(html, 'html.parser')
```

```python
cnt = bs.select('div.search-result-section > span > span')
'''
for x in cnt:
  print(x.getText().replace(',',""))
  print(type(x.getText()))
'''
cnt = int(cnt[0].get_text().replace(',',''))
print(type(cnt))
```

## 페이지 개수 만들기

```python
pageno = int(cnt / 10 + 0.99)
print(pageno)
```

## 기사링크 가져오기

```python
links = []

for i in range(0, 3):
  url = 'https://search.hani.co.kr/search/newslist?searchword={}&page={}'.format(keyword, i+1)
  print(url)
  html = requests.get(url).text
  bs = BeautifulSoup(html, 'html.parser')
  linktag = bs.select('article > a')
  for tag in linktag:
    links.append(tag['href'])

print(links)
```

## 기록할 텍스트 파일 생성

```python
output_file  = open(keyword + '.txt', 'w', encoding='utf-8')

for link in links:
  html = requests.get(link).text
  print(html)
  bs = BeautifulSoup(html, 'html.parser')
  articles = bs.select('h3.ArticleDetailView_title__9kRU_')
  print(articles)
  for article in articles:
    text = article.get_text(strip=True)
    output_file.write(text + '\n')

output_file.close()
```
