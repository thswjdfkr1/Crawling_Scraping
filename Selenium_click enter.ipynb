```python
# ÌÇ§Î≥¥Îìú ÏóîÌÑ∞

options = webdriver.ChromeOptions()
options.add_argument('--headless')  #Ìó§ÎìúÎ¶¨Ïä§
options.add_argument('--no-sandbox')  # ÏÉåÎìúÎ∞ïÏä§ ÎπÑÌôúÏÑ±Ìôî
options.add_argument('--disable-dev-shm-usage')  # /dev/shm ÎπÑÌôúÏÑ±Ìôî
options.add_argument('--disable-dev-shm-usage')  # ÏÇ¨Ïö©Ïûê Îç∞Ïù¥ÌÑ∞ ÎîîÎ†âÌÜ†Î¶¨ ÏÑ§Ï†ï

driver = webdriver.Chrome(options=options)

try:
    driver.get("https://www.naver.com")

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#query")))

    search_box = driver.find_element(By.CSS_SELECTOR, "#query")
    search_box.send_keys("ÌîÑÎ°úÏïºÍµ¨")

    search_button = driver.find_element(By.CSS_SELECTOR, "#search_btn")
    search_button.click()

finally:
    driver.quit()
```
