```python
## 뉴스 타이틀 가져오기
```

```python
# url 넘겨주기
news_q = requests.get('https://search.naver.com/search.naver?sm=tab_hty.top&where=news&ssc=tab.news.all&query=%EB%91%90%EC%82%B0%EB%B2%A0%EC%96%B4%EC%8A%A4&oquery=entksqpdjtm&tqi=jb%2BsRwqo15wssLDAeP4ssssstQN-181332&ackey=7sijo5x2')
news_q.text
```

```python
# 문자열 데이터 html화
soup2 = bs(news_q.text, 'lxml')
```

```python
# 원하는 뉴스 제목 찾기
soup2.select('div.news_contents > a')[1].text
title = soup2.select('div.news_contents > a')
title
```

```python
# 뉴스제목 리스트에 담기
news_title = []

for i in range(len(title)):
    news_title.append(title[i].text)
print(news_title)
```

```python
# 원하는 뉴스 제목 찾기
titles = soup2.select('a.news_tit')
```

```python
# 뉴스제목 리스트에 담기
news_title = []

for i in range(len(titles)):
    news_title.append(titles[i].text)
print(news_title)
```

```python
## 멜론 차트 수집
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
user_agent = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36'}

res = requests.get('https://www.melon.com/chart/index.htm', headers = user_agent)
res
```

```python
res.text
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
# 노래 제목 수집
title = soup.select('div.ellipsis.rank01 > span > a ')
title
```

```python
titles_list = []

for i in title:
    titles_list.append(i.text)
```

```python
len(titles_list)
```

```python
## 가수 이름 수집
```

```python
singer = soup.select('div.ellipsis.rank02 > span ')
singer
```

```python
singers_list = []

for i in singer:
    singers_list.append(i.text)
```

```python
len(singers_list)
```

```python
# 데이터 프레임화
import pandas as pd

melon_top100 = pd.DataFrame({'노래제목': titles_list, '가수명':singers_list})
melon_top100
```

```python
melon_top100.to_csv('melon_top100.csv',index=False)
```
