```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time
import pandas as pd
```

```python
# 드라이버 경로 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options = options)
```

```python
try:
  driver.get('https://www.koreabaseball.com/Player/RegisterALL.aspx')

  player_name = []
  for i in range(1,10):
    catchers = 'td:nth-child(5) > ul > li:nth-child({})'.format(i)
    catchers = driver.find_elements(By.CSS_SELECTOR, catchers)
    for players in catchers:
      player_name.append(players.text)

  for i in range(1,10):
    infielders = 'td:nth-child(6) > ul > li:nth-child({})'.format(i)
    infielders = driver.find_elements(By.CSS_SELECTOR, infielders)
    for players in infielders:
      player_name.append(players.text)

  for i in range(1,10):
    outfielders = 'td:nth-child(7) > ul > li:nth-child({})'.format(i)
    outfielders = driver.find_elements(By.CSS_SELECTOR, outfielders)
    for players in outfielders:
      player_name.append(players.text)

  with open('players.csv', 'w', newline='',encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    writer.writerow(['player_name'])
    for name in player_name:
      writer.writerow([name])

finally:
  driver.quit()
```

```python
# 2군 포함

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# driver = webdriver.Chrome(options = options)
```

```python
driver = webdriver.Chrome(options = options)

try:
  player_name = []

  # 1군
  driver.get('https://www.koreabaseball.com/Player/RegisterALL.aspx')
  wait = WebDriverWait(driver, 10)

  wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'td:nth-child(5) > ul')))

  positions = ['td:nth-child(5) > ul > li',
               'td:nth-child(6) > ul > li',
               'td:nth-child(7) > ul > li']
  for pos in positions:
    elements = driver.find_elements(By.CSS_SELECTOR, pos)
    for element in elements:
      player_name.append(element.text)

  # 2군
  driver.get('https://www.koreabaseball.com/Futures/Player/Register.aspx')

  team_ids = ['HH','LG','WO','SK','OB','SM','NC','SS','LT','KT','HT']
  for team_id in team_ids:
    element = driver.find_element(By.XPATH, f"//li[@data-id='{team_id}']/a")
    element.click()
    time.sleep(7)

    positions = ['table:nth-child(4)', 'table:nth-child(5)','table:nth-child(6)']
    for pos in positions:
      player = driver.find_elements(By.CSS_SELECTOR, f'{pos} > tbody > tr > td:nth-child(2)')
      for player in player:
        player_name.append(player.text)

finally:
  driver.quit()

with open('all2_players.csv', 'w', newline='',encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    writer.writerow(['player_name'])
    for name in player_name:
      writer.writerow([name])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
columns = ['Rank','Name','Team','WAR','G','oWAR','dWAR','PA','ePA','AB','R','H',
           '2B','3B','HR','TB','RBI','SB','CS','BB','HP','IB','SO','GDP','SH',
           'SF','AVG','OBP','SLG','OPS','R/ePA','wRC+','WAR']
data = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

for year in range(2002, 2025):
  url = f'https://statiz.sporki.com/stats/?m=main&m2=batting&m3=default&so=WAR&ob=DESC&year={year}&sy=&ey=&te=&po=&lt=10100&reg=R&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=50&ph=&hs=&us=&na=&ls=&sf1=&sk1=&sv1=&sf2=&sk2=&sv2='
  response = requests.get(url, headers = headers)
  soup = BeautifulSoup(response.text, 'html.parser')
  time.sleep(5)

  # 데이터가 있는 테이블을 찾아 각 행을 순회
  rows = soup.find('table').find('tbody').find_all('tr')
  for row in rows:
    cols = row.find_all('td')
    if len(cols) > 1: # 헤더나 공백이 아닌 겨우
      col_data  = []
      for col in cols:
        col_data.append(col.get_text().strip())
      data.append(col_data)

df = pd.DataFrame(data, columns = columns)

df.to_csv('kbo_players_stats_2002_to_2025.csv', index=False, encoding='utf-8-sig')
```
