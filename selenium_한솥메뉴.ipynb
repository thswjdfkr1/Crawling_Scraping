```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
#    
driver = webdriver.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')

# bnt = driver.find_element(By.CLASS_NAME, 'c_05')
# btn.click()
#   
while True:
    try:
        btn = driver.find_element(By.CLASS_NAME, 'c_05')
        btn.click()
        print('"" ')
        time.sleep(2)
        
    except:
        print('   ')
        break
    
#   
menu = driver.find_elements(By.CLASS_NAME, 'h.fz_03')
print(menu[0].text)
print(len(menu))

# 1  
price = driver.find_elements(By.CLASS_NAME, 'item-price > strong')
price[0].text

#  ,   
menu_list = []
price_list = []
for i in range(len(menu)):
    name = menu[i].text
    menu_list.append(name)
for j in price:
    price_list.append(j.text)
for i in range(len(menu_list)):
    print(menu_list[i],': ', price_list[i])
menu_price = pd.DataFrame({'':menu_list, '':price_list})
# menu_price
    
menu_price.to_csv('.csv', encoding='euc-kr')
driver.close()
    
```
