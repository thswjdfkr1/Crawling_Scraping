```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd

from selenium.webdriver.chrome.options import Options
```

```python
# 헤드리스 모드(브라우저 창 없이)
chrome_options = Options()
chrome_options.add_argument('--headless')
driver = webdriver.Chrome(options = chrome_options)

keyword = input('검색어를 입력하세요 >')
driver.get(f'https://www.youtube.com/results?search_query={keyword}')

# 태그찾기
title = driver.find_elements(By.CSS_SELECTOR, '#video-title > yt-formatted-string')
print('첫번째 제목:', title[2].text)

# 제목 리스트에 담기
title_list = []
for i in title:
    title_list.append(i.text)
print('전체 제목:', title_list, '\n', '제목 길이:',len(title_list))

# 스크롤 하강
for i in range(5):
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(2)

# 태그 다시 수집
# titles = driver.find_elements(By.CSS_SELECTOR, '#video-title > yt-formatted-string')
title_url = driver.find_elements(By.CSS_SELECTOR, 'a#video-title')

titles_list = []
for i in title_url:
    # titles_list.append(i.text)
    titles_list.append(i.get_attribute('title'))
print('스크롤 후 제목:', titles_list, '\n', '스크롤 후 제목 길이:',len(titles_list))

# url 수집
print('첫번쨰 url:', title_url[0].get_attribute('href'))

url_list = []
for i in title_url:
    url_list.append(i.get_attribute('href'))
print('총 주소 길이:', len(url_list))

url_df = pd.DataFrame({'제목':titles_list,'주소': url_list})
display(url_df)

url_df.to_csv('유튜브_두산베어스.csv', encoding = 'utf-8-sig', index=False)
driver.close()
```
